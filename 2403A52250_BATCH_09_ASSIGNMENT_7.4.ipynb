```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
```

```python
D1="I like machine learning"
D2="I like deep learning"
D3="I like playing tennis"
D4="I like playing cricket"
D5="I enjoy data science"
D6="I enjoy artificial intelligence"
D7="I love natural language processing"
D8="I love computer vision"
D9="I prefer online education"
D10="I prefer classroom learning"
D11="I like watching movies"
D12="I like watching documentaries"
D13="I enjoy playing football"
D14="I enjoy playing basketball"
D15="I like reading books"
D16="I like reading novels"
D17="I love listening to music"
D18="I love classical music"
D19="I enjoy traveling abroad"
D20="I enjoy exploring nature"
D21="I like healthy food"
D22="I like spicy food"
D23="I enjoy coding in python"
D24="I enjoy coding in java"
D25="I like cloud computing"
D26="I like cybersecurity"
D27="I love robotics projects"
D28="I love automation systems"
D29="I enjoy big data analytics"
D30="I enjoy data visualization"

documents = [
D1,D2,D3,D4,D5,D6,D7,D8,D9,D10,
D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,
D21,D22,D23,D24,D25,D26,D27,D28,D29,D30
]
```

```python
vectorizer = CountVectorizer(stop_words="english")
bow = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

df_bow = pd.DataFrame(
    bow.toarray(),
    columns=feature_names,
    index=[f"D{i}" for i in range(1,31)]
)

print("Bag of Words Representation:")
print(df_bow)
```

cosine similarity

```python
cosine_sim_matrix = cosine_similarity(df_bow)

df_cosine = pd.DataFrame(
    cosine_sim_matrix,
    index=[f"D{i}" for i in range(1,31)],
    columns=[f"D{i}" for i in range(1,31)]
)

print("\nCosine Similarity Matrix:")
print(df_cosine)
```

jacard similarity

```python
binary_bow = (df_bow > 0).astype(int)

num_documents = len(documents)
jaccard_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        vec_i = binary_bow.iloc[i]
        vec_j = binary_bow.iloc[j]

        intersection = np.sum(np.logical_and(vec_i, vec_j))
        union = np.sum(np.logical_or(vec_i, vec_j))

        if union == 0:
            jaccard_sim_matrix[i, j] = 0
        else:
            jaccard_sim_matrix[i, j] = intersection / union

df_jaccard = pd.DataFrame(
    jaccard_sim_matrix,
    index=[f"D{i}" for i in range(1,31)],
    columns=[f"D{i}" for i in range(1,31)]
)

print("\nJaccard Similarity Matrix:")
print(df_jaccard)
```

```python
# Download if not present
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab') # Added to resolve LookupError

def document_wordnet_similarity(doc1, doc2):
    tokens1 = word_tokenize(doc1.lower())
    tokens2 = word_tokenize(doc2.lower())

    synsets1 = [s for token in tokens1 for s in wordnet.synsets(token)]
    synsets2 = [s for token in tokens2 for s in wordnet.synsets(token)]

    if not synsets1 or not synsets2:
        return 0.0

    max_similarities = []

    for s1 in synsets1:
        max_sim_for_s1 = 0
        for s2 in synsets2:
            sim = s1.path_similarity(s2)
            if sim is not None and sim > max_sim_for_s1:
                max_sim_for_s1 = sim
        max_similarities.append(max_sim_for_s1)

    return np.mean(max_similarities) if max_similarities else 0.0


wordnet_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        if i == j:
            wordnet_sim_matrix[i, j] = 1.0
        else:
            wordnet_sim_matrix[i, j] = document_wordnet_similarity(documents[i], documents[j])

df_wordnet = pd.DataFrame(
    wordnet_sim_matrix,
    index=[f"D{i}" for i in range(1,31)],
    columns=[f"D{i}" for i in range(1,31)]
)

print("\nWordNet Similarity Matrix:")
print(df_wordnet)
```
